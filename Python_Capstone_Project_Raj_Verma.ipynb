# **Part 1: Basic Operations and Arithmetic**

This section covers fundamental Python operations including variable assignment, arithmetic operations, and basic calculations.

### **Question 1 – Score Board Operator**

**Concept / Problem Statement:**  
An operator working behind the scoreboard of an inter-cohort competition needs to calculate and display the total points scored by Team London. The team scored 45 points in the first round and 32 points in the second round.

```python
# Define the points scored in each round
round_1 = 45
round_2 = 32

# Calculate the total points gained by Team London
london_points = round_1 + round_2

# Print the total points
print(f"Team London scored a total of {london_points} points")
```

### **Question 2 – Nutrition Bar Ingredient Chart**

**Concept / Problem Statement:**  
You are given the percentage composition of ingredients in a nutrition bar. The bar contains 20% raisins, 30% almonds, and the rest are apricots. Calculate each ingredient's percentage.

```python
# Total percentage must equal 100
total_percentage = 100

# Calculate the percentage of each ingredient
raisins = 20
almonds = 30
apricots = total_percentage - raisins - almonds

# Display the ingredient composition
print(f"Nutrition Bar Composition:")
print(f"Raisins: {raisins}%")
print(f"Almonds: {almonds}%")
print(f"Apricots: {apricots}%")
```

### **Question 3 – Measuring Accuracy of a Model**

**Concept / Problem Statement:**  
Calculate the accuracy of a machine learning model. If the model made 80 correct predictions out of 100 total cases, what is the accuracy?

```python
# Store the number of correct predictions and total cases
correct_predictions = 80
total_cases = 100

# Calculate accuracy as a decimal
accuracy = correct_predictions / total_cases

# Print the result
print(f"Model Accuracy: {accuracy}")
print(f"Model Accuracy (Percentage): {accuracy * 100}%")
```

### **Question 4 – Arithmetic Operations**

**Concept / Problem Statement:**

### **Question 16 – Fibonacci Series**

**Concept / Problem Statement:**  
Generate Fibonacci numbers up to a given limit.

```python
# Define the number of terms
n_terms = 6

# Initialize first two terms
a, b = 0, 1

# Print Fibonacci series
print(f"First {n_terms} Fibonacci numbers:")
for _ in range(n_terms):
    print(b, end=" ")
    a, b = b, a + b
print()  # New line
```

# **Part 6: Exception Handling and OOP**

This section covers exception handling for error management and basic object-oriented programming concepts.

### **Question 26 – Exception Handling**

**Concept / Problem Statement:**  
Handle runtime errors using try-except blocks to prevent program crashes.

```python
# Demonstrate exception handling
try:
    # Attempt to divide by zero
    result = 10 / 0
    print(result)
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")
    print("Exception handled successfully.")

# Program continues to run
print("Program continues after error handling.")
```

### **Question 30 – Basic OOP: Class and Object**

**Concept / Problem Statement:**  
Create a class and object to demonstrate object-oriented programming principles.

```python
# Define a Person class
class Person:
    """A simple class representing a person"""

    def __init__(self, name, age):
        """Constructor to initialize person attributes"""
        self.name = name
        self.age = age

    def introduce(self):
        """Method to introduce the person"""
        print(f"Hello! My name is {self.name} and I am {self.age} years old.")

    def birthday(self):
        """Method to increment age"""
        self.age += 1
        print(f"Happy Birthday! {self.name} is now {self.age} years old.")

# Create an object of Person class
person1 = Person('Raj', 27)

# Call methods
person1.introduce()
person1.birthday()
```
